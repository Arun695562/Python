```python
from sklearn.linear_model import LinearRegression
import pandas as pd

data = pd.DataFrame({
    "Study_hours": [1, 2, 3, 4, 5],
    "Marks": [30, 40, 50, 60, 70]
})


X = data[["Study_hours"]]
y = data["Marks"]

model = LinearRegression()
model.fit(X, y)

predict = model.predict([[6]])
predict = model.predict(pd.DataFrame([[6]], columns=X.columns))
print(predict)

print(predict)
Predicted = model.predict(X)
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, Predicted, color = "red",label = "Predicted value")
plt.legend()
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression")
plt.grid(True)
plt.show()
```
